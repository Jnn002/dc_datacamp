## Difference between Line and scatter plots

##### Scatter plots
- Each plot is and independent observation

#### Line plots
- Each plot point represents the same 'thing' typically tracked over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Scatter plots

```python
students = pd.read_csv('student-alcohol-consumption.csv', index_col= 0)
students.head()
```

```python
sns.relplot(x="G1", y="G3", 
            data=students,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row = 'famsup',
            row_order = ['yes', 'no'],
            size = 'absences',    
            hue = 'absences', 
            alpha = 1        
            )

# Show plot
plt.show()
```

```python
cars = pd.read_csv('mpg.csv')


sns.relplot(data = cars, kind = 'scatter', x = 'acceleration', y = 'mpg', style = 'origin', hue = 'origin')
```

#### Line plots

```python
# Create line plot
sns.relplot(data = cars, kind = 'line', x = 'model_year', y = 'mpg')

# Show plot
plt.show()
```
