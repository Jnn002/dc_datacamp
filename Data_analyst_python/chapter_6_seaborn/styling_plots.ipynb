##### Changing the figure style
- Figure 'style' includes background and axes
- Preset options: 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('student-alcohol-consumption.csv', index_col= 0)

sns.set_style('whitegrid')
#sns.set_palette('RdBu') 
sns.set_palette(['#E94E77', '#36ADA4']) # We can also use HEX codes
sns.set_context('notebook')

graph = sns.catplot(data = students, kind = 'bar', 
                x = 'study_time', 
                y = 'G3',
                order = ["<2 hours", 
                        "2 to 5 hours", 
                        "5 to 10 hours", 
                        ">10 hours"],
                errorbar= None,
                hue = 'sex',
                col = 'location'
                #palette = ['r', 'k']
                )

# Add a title
graph.figure.suptitle('Average Final Grade by Study Time', y = 1.05)
graph.set_titles('Location: {col_name}')

# Add labels
graph.set_axis_labels(xlabel = 'Study Time', ylabel ='Final Grade')   # We use the set_axis_labels method to add labels
#graph.set(xlabel = 'Study Time', ylabel ='Final Grade')               # We can also use the set method to add labels

graph.set_xticklabels(rotation = 5)

plt.show()
```
