## Difference between Line and scatter plots

##### Scatter plots
- Each plot is and independent observation

#### Line plots
- Each plot point represents the same 'thing' typically tracked over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Scatter plots

```python
students = pd.read_csv('student-alcohol-consumption.csv', index_col= 0)
students.head()
```

```python
sns.relplot(x="G1", y="G3", 
            data=students,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row = 'famsup',
            row_order = ['yes', 'no'],
            size = 'absences',    
            hue = 'absences', 
            alpha = 1        
            )

# Show plot
plt.show()
```

```python
cars = pd.read_csv('mpg.csv')


sns.relplot(data = cars, kind = 'scatter', x = 'acceleration', y = 'mpg', style = 'origin', hue = 'origin')
plt.show()
```

#### Line plots

```python
# Create line plot
sns.relplot(data = cars, kind = 'line', 
            x = 'model_year', 
            y = 'horsepower', 
            errorbar = None,
            style = 'origin',
            hue = 'origin',
            dashes = False,
            markers = True
            )

# Show plot
plt.show()
```

### Count and bar plots

##### This are categorical type plot
- Involves a categorical variable
- Comparisons between groups


##### Catplot () is used to create categorical plots

```python
students = pd.read_csv('student-alcohol-consumption.csv', index_col= 0)
students.head(2)
```

```python
sns.catplot(data = students, kind = 'bar', 
                x = 'study_time', 
                y = 'G3',
                order = ["<2 hours", 
                        "2 to 5 hours", 
                        "5 to 10 hours", 
                        ">10 hours"],
                errorbar= None,
                hue = 'sex',
                palette = ['r', 'k']
                )

plt.title('Average Final Grade by Study Time')
plt.show()
```

```python
# Boxplot

# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(data = students, kind = 'box', x = 'study_time', y = 'G3', order = study_time_order)

# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(data = students, kind = 'box', 
            x = 'internet', 
            y = 'G3', 
            hue = 'location', 
            sym = '',
            whis = [10, 90]
            )

# Show plot
plt.show()
```

#### Point plots

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=students,
            kind="point",
            hue="school",
            errorbar = None,
            estimator = median)

# Show plot
plt.show()
```
