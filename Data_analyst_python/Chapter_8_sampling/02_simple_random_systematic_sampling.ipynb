### Simple sampling


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
coffee = pd.read_feather('datasets/coffee_ratings_full (1).feather')
coffee.head()
```

```python
# Simple sampling
# With random state set to 42, the same random sample will be selected each time

coffee_sample = coffee.sample(n=20, random_state = 42)
coffee_sample.head()
```

```python
# Systematic sampling
sample_size = 20
pop_size = coffee.shape[0]

interval = pop_size // sample_size  # 66, every 66th row will be selected

# We shoulf make sure that the sample is not biased
coffee.iloc[::interval].head()
```

```python
shuffled = coffee.sample(frac=1)
shuffled = shuffled.reset_index(drop=True).reset_index()

shuffled.plot(x='index', y='category_two_defects', kind='scatter')
plt.show()
```

#### Stratified sampling

```python
top_counted_countries = ['Mexico', 'Guatemala', 'Colombia', 'Brazil', 'Taiwan']

top_counted_subset = coffee['country_of_origin'].isin(top_counted_countries)

coffee_ratings_top = coffee[top_counted_subset]

# sampling
coffee_ratings_samp = coffee_ratings_top.sample(frac = 1, random_state= 2022)

coffee_ratings_samp['country_of_origin'].value_counts(normalize=True)
```

#### Proportional stratified sampling

```python
coffee_ratings_strat = coffee_ratings_top.groupby('country_of_origin')\
                        .sample(frac = 1, random_state= 2022)
                        
coffee_ratings_strat['country_of_origin'].value_counts(normalize=True)
```

```python
coffee_ratings_eq = coffee_ratings_top.groupby('country_of_origin')\
                        .sample(n = 15, random_state= 2022)
                        
coffee_ratings_eq['country_of_origin'].value_counts(normalize=True)
```

```python
coffee_ratings_weight = coffee_ratings_top
condition = coffee_ratings_weight['country_of_origin'] == 'Taiwan'

coffee_ratings_weight['weight'] = np.where(condition, 2, 1)

coffee_ratings_weight = coffee_ratings_weight.sample(frac = 0.1, weights= 'weight')

coffee_ratings_weight['country_of_origin'].value_counts(normalize=True)
```

```python
coffee_ratings_weight[coffee_ratings_weight['weight'] == 2].head()
```

##### Stratified samplings vs Cluster sampling

- Stratified sampling
  - Split the population into subgroups
  - Use simple random sampling on every subgroup
- Cluster Sampling
  - Use simple random sampling to pick some subgroups
  - Use simple random sampling on only those subgrous

SO, the main benefit of cluster sampling over stratifed sampling is that you can save time and money by not including every subgroup in your sample
