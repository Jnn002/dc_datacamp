```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

spotify = pd.read_feather('datasets/spotify_2000_2020.feather')


spotify.head()
```

```python
spotify_sample = spotify.sample(1000)
spotify_sample
```

```python
mean_spotify_population = spotify['duration_minutes'].mean()
mean_spotify_sample = spotify_sample['duration_minutes'].mean()

print(f'Mean population: {"%.3f" % mean_spotify_population}\nMean sample: {"%.3f" % mean_spotify_sample}')
```

### Why we have a good quality sample?

We need to have a good quality sample without bias to ensure a high quaiity analisis

```python
sns.histplot(data = spotify, bins = 1, binwidth= 0.01, x = 'acousticness', kde = True)
plt.show()
```
