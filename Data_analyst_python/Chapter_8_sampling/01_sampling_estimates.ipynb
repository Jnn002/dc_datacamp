```python
import pandas as pd

spotify = pd.read_feather('datasets/spotify_2000_2020.feather')


spotify.head()
```

```python
spotify_sample = spotify.sample(1000)
spotify_sample
```

```python
mean_spotify_population = spotify['duration_minutes'].mean()
mean_spotify_sample = spotify_sample['duration_minutes'].mean()

print(f'Mean population: {"%.3f" % mean_spotify_population}\nMean sample: {"%.3f" % mean_spotify_sample}')
```
