```python
import pandas as pd
import numpy as np

# Create a dictionary with data about dogs
data = {
    'Name': ['Bella', 'Lucy', 'Max', 'Charlie', 'Buddy', 'Rocky', 'Molly', 'Daisy', 'Bailey', 'Lola','Bella', 'Lucy', 'Max'],
    'Age': [3, 5, 2, 4, 6, 1, 7, 3, 5, 2, 3, 5, 2],
    'Breed': ['Labrador', 'Poodle', 'Bulldog', 'Beagle', 'Pug', 'Boxer', 'Dachshund', 'Shih Tzu', 'Husky', 'Chihuahua', 'Labrador', 'Poodle', 'Bulldog'],
    'Weight': [22, 18, 25, 20, 10, 30, 12, 8, 28, 6, 22, 18, 25],
    'Date': pd.date_range(start='2023-01-01', periods=13, freq='D').tolist()
}
# Create a DataFrame from the dictionary
df_dogs = pd.DataFrame(data)
```

```python
# We can use the value_counts() method to count the number of occurrences of each breed in the DataFrame
print(df_dogs['Breed'].value_counts())
# get percentage of each breed
print(df_dogs['Breed'].value_counts(normalize=True) * 100)
```

```python
# Dropping duplicates
vet_visits = df_dogs
vet_visits.drop_duplicates(subset=['Name', 'Breed'], keep='first', inplace=True)
unique_dogs = vet_visits
unique_dogs
```
