```python
import pandas as pd
import matplotlib.pyplot as plt

dogs = pd.read_csv('dogs.csv')
dogs['height'].hist(bins = 30)
plt.show()
```

```python
avg_weight_by_breed = dogs.groupby('breed')['weight'].mean()

plt.bar(avg_weight_by_breed.index, avg_weight_by_breed)
plt.xlabel('Breeds')
plt.xticks(rotation = 90)
plt.title('Average weight by breed')
plt.show()
```

```python
dogs.plot(kind = 'scatter', x = 'height', y = 'weight')
plt.show()
```
