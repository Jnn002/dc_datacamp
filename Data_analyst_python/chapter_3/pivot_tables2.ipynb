```python
import pandas as pd

dogs = pd.read_csv('dogs.csv')

dogs.head()
```

```python
# CONVERTING 'date_of_birth' TO DATETIME
dogs['date_of_birth'] = pd.to_datetime(dogs['date_of_birth'])

# ADDING A NEW COLUMN FOR YEAR OF BIRTH
dogs['year_birth'] = dogs['date_of_birth'].dt.year

dogs.head()
```

```python
# Get the year with more dogs born
year_with_most_dogs = dogs['year_birth'].value_counts()
year_with_most_dogs
```
