```python
import pandas as pd

# create a dataframe with sample data about dogs
dogs = pd.DataFrame({
    'Name': ['Buddy', 'Bella', 'Charlie', 'Max', 'Lucy', 'Daisy', 'Rocky', 'Molly', 'Duke', 'Luna'],
    'Age': [3, 5, 2, 4, 6, 1, 7, 8, 3, 2],
    'Breed': ['Golden Retriever', 'Labrador', 'Poodle', 'Bulldog', 'Beagle', 'Pug', 'Boxer', 'Dachshund', 'Shih Tzu', 'Chihuahua'],
    'Color': ['Gold', 'Black', 'White', 'Brown', 'Black', 'Tan', 'Brown', 'White', 'White', 'Tan'],
    'weight': [29, 10, 10, 22, 25, 15, 15, 15, 20, 10],
    'height': [59, 35, 20, 20, 35, 22, 27, 32, 44, 60]
})
```

```python
dogs_srt = dogs.set_index(['Breed', 'Color']).sort_index()
dogs_srt
```

```python
# select all dogs with breed between Boxer and Labrador
# If we need to slice at two levels of a multiindex, we need to use the loc accessor

dogs_srt.loc['Boxer':'Chihuahua']
dogs_srt.loc[('Boxer', 'Brown'):('Dachshund', 'White')]
```

```python
# We can slice columns as well
# select all dogs with breed between Boxer and Labrador and only the Name and Age columns
dogs_srt.loc[('Boxer', 'Brown'):('Dachshund', 'White'), ['Name', 'Age']]
```

```python
# And as we see before we are able to slice with position as well

# select all dogs with breed between Boxer and Labrador and only the Name and Age columns
dogs.iloc[6:9, 0:3]
```
