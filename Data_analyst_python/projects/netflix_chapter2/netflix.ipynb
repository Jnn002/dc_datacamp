```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Display most frecuente Movie duration in 1990s, variable duration 
# Filter the data so that only movies released in the 1990s are in the DataFrame
filter_type = netflix_df[netflix_df['type'] == 'Movie']
filter_range= filter_type[(filter_type["release_year"] >= 1990) & (filter_type["release_year"] < 2000)]


plt.hist(filter_range["duration"], bins=10, edgecolor='black', color = 'purple')
plt.xlabel('Duration in mins')
plt.ylabel('Number of Movies')
plt.title('Movie Duration in the 90s')
plt.show()

# Imprimir la duracion promedio de las peliculas en la decada de los 90s
print(int(filter_range["duration"].mean()))
```

```python
# DIsplay the number of short movies of action in the 1990s
count_action_movies = filter_range[filter_range['genre'] == 'Action']

final_count = len(count_action_movies[count_action_movies["duration"] < 90])
print(f'The number of actions movies that are less than 90 mins in 1990s is: {final_count}')
```
