```python
# Random numbers
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
tails = [0]

for x in range(10):
    coin = np.random.randint(0, 2)
    tails.append(tails[x] + coin)
print(tails)
```

```python
import matplotlib.pyplot as plt
# Initialization
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

# Import matplotlib.pyplot as plt


# Plot random_walk
plt.plot(random_walk)
plt.xlabel('Number of steps')
plt.ylabel('Random walk')
plt.title( 'Random walk with dice in Empire State Building')
# Show the plot
plt.show()
```
