```python
# Random numbers
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
tails = [0]

for x in range(10):
    coin = np.random.randint(0, 2)
    tails.append(tails[x] + coin)
print(tails)
```

```python
import matplotlib.pyplot as plt
# Initialization
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

# Import matplotlib.pyplot as plt


# Plot random_walk
plt.plot(random_walk)
plt.xlabel('Number of steps')
plt.ylabel('Random walk')
plt.title( 'Random walk with dice in Empire State Building')
# Show the plot
plt.show()
```

```python
# Distribution of random walk

np.random.seed(123)
final_tails = []

for x in range(80000):
    tails = [0]
    for x in range(10):
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])
plt.hist(final_tails, bins = 10, edgecolor = 'grey')
plt.xticks(range(0,11))
plt.xlabel('Number of tails')
plt.ylabel('Number of simulations')
plt.show()
```

```python
# Simulating multiple walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```
