```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

# Adding more data 

year = year + [2015, 2020, 2030]
pop = [1.0, 1.262, 1.650] + pop
# The order matters, the first element in the list will be the first element in the plot
# but we could use the sort method to sort the data

plt.plot(year, pop)
# Adding labels
# We could assing the value of a variable to the label
xlab = 'year'
plt.xlabel(xlab)
# We could also use the string directly
plt.ylabel('population (billion)')
plt.title('world population projection')
# Adding ticks
plt.yticks([0, 2, 4, 6, 8]
            , ['0', '2B', '4B', '6B', '8B'])

plt.show()
```
